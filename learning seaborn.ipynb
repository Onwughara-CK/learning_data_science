```python
import seaborn as sns
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
sns.distplot(tips['size'],kde = True,bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.pairplot(tips,hue = 'sex',diag_kind='hist')
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
#categorical data plots
sns.barplot(x='sex',y='tip',data=tips,estimator = np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue ='time')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue ='sex')
```

```python
flights.head(3)
```

```python
sns.boxplot(x='month',y='passengers',data=flights)
```

```python
fpt = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fpt
```

```python
sns.heatmap(fpt,cmap='coolwarm',lw=2,linecolor='black')
```

```python
sns.clustermap(fpt,cmap='coolwarm')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_diag(sns.distplot)
g.map_lower(sns.swarmplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('title')
```

```python
plt.subplot(121)
plt.plot(x,y,'g')
plt.title('title')

plt.subplot(122)
plt.plot(y,x,'y')
plt.title('title')


```

```python
#OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)


axes.set_title('Title')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.0,0.0,1,1])
axes2 = fig.add_axes([0.6,0.1,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('bigger')
axes2.set_title('smaller')
```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[0].set_title('title1')
axes[1].plot(y,x)
axes[1].set_title('title2')
plt.tight_layout()
```

```python
fig,axes = plt.subplots(1,2,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
axes
```

```python
fig.savefig('my_pic.png',dpi=200)
```

```python
fig = plt.figure(figsize = (8,2))
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label = 'x-square')
axes.plot(x,x**3, label = 'x-cube')

axes.legend(loc = 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,lw =2,ls='--',color = 'black')

#how to show a plot between 0 and 1 on the x-axis

ax.set_xlim([0,4])
ax.set_ylim([0,2])
```
